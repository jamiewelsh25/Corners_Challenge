```python
# Group by 'Year'
grouped_stats = df.groupby('Year')['Total_Corners'].describe()

# Calculate mode separately
grouped_stats['Mode'] = df.groupby('Year')['Total_Corners'].apply(lambda x: x.mode().iloc[0])

# Display the resulting statistics
print(grouped_stats[['mean', '50%', 'Mode', '25%', '75%', 'std']])
```

```python
from scipy.stats import ks_2samp
import itertools

# Assuming df is your DataFrame
# Also assuming you have 'Total_Corners' and 'Year' columns

# Set alpha level
alpha = 0.05

# Get unique years
unique_years = df['Year'].unique()

# Iterate through unique pairs of years
for year1, year2 in itertools.combinations(unique_years, 2):
    
    # Extract Total Corners for the two years
    sample1 = df[df['Year'] == year1]['Total_Corners']
    sample2 = df[df['Year'] == year2]['Total_Corners']
    
    # Perform Kolmogorov-Smirnov test
    _, p_value = ks_2samp(sample1, sample2)
    
    # Print the results
    print(f'Year {year1} vs. Year {year2}: p-value = {p_value}')
    
    # Check for statistical significance
    if p_value < alpha:
        print('   Statistically significant difference')
    else:
        print('   No statistically significant difference')
```

```python
from scipy.stats import ks_2samp

# Set alpha level
alpha = 0.05

# Get unique teams
unique_teams = set(df['HomeTeamId'])

# Initialize a list to collect team IDs with significant differences
teams_with_significant_difference = []

# Iterate through each team
for team in unique_teams:
    
    # Extract Total Corners for the current team and all other teams
    sample_team = df[(df['HomeTeamId'] == team)]['Total_Corners']
    sample_other_teams = df[(df['HomeTeamId'] != team)]['Total_Corners']
    
    # Perform Kolmogorov-Smirnov test
    _, p_value = ks_2samp(sample_team, sample_other_teams)
    
    # Check for statistical significance
    if p_value < alpha:
        teams_with_significant_difference.append(team)

# Print the fraction of teams with significant differences
fraction_teams_with_difference = len(teams_with_significant_difference) / len(unique_teams)
print(f'Fraction of home teams with significant differences: {fraction_teams_with_difference:.2%}')

# Optionally, print the team IDs with significant differences
print('Home teams with significant differences:', teams_with_significant_difference)

```

## Feature Engineering

```python
# Sort the DataFrame by date or any other appropriate column
df.sort_values(by='Date', inplace=True)  # Replace 'Date' with the actual date column in your DataFrame

# Function to calculate mean corners for the last five games or fill with league average
def calculate_mean_corners(df, team_id_col, corners_col, league_id_col, new_col_name):
    for index, row in df.iterrows():
        
        team_id = row[team_id_col]
        league_id = row[league_id_col]
        
        # Filter previous games for the same team
        previous_games = df[(df[team_id_col] == team_id) & (df['Date'] < row['Date'])].tail(5)
        
        
        
        # Calculate mean for the last five games
        last_five_mean = previous_games[corners_col].mean()
        
        # If there are fewer than five previous games, use league average
        if len(previous_games) > 4:
            df.at[index, new_col_name] = last_five_mean
        else:
            if len(df[(df[league_id_col] == league_id) & (df['Date'] < row['Date'])][corners_col].values) > 4:
                league_avg = df[(df[league_id_col] == league_id) & (df['Date'] < row['Date'])][corners_col].values.mean()
                
                df.at[index, new_col_name] = league_avg
            else:
                df.at[index, new_col_name] = np.nan

# Calculate mean Home_Corners for the last five games or fill with league average
calculate_mean_corners(df, 'HomeTeamId', 'Total_Corners', 'LeagueId', 'Mean_Home_Corners_Prev5')

# Calculate mean Away_Corners for the last five games or fill with league average
calculate_mean_corners(df, 'AwayTeamId', 'Total_Corners', 'LeagueId', 'Mean_Away_Corners_Prev5')

# Reset the DataFrame index
df.reset_index(drop=True, inplace=True)

# Calculate the rolling mean for 'Total Corners' for each league
df['League_Avg_Corners'] = df.groupby('LeagueId')['Total_Corners'].transform(lambda x: x.rolling(window=len(x), min_periods=50).mean().shift(1))


def expected_result_prob(home_score,away_score, weight = 400, home_field_advantage = 100):
    # calculate the expected probabilities for a match between two teams
    diff_h_a = home_score - away_score + home_field_advantage #here we can add home field advantage

    we = (1/(10 ** (-diff_h_a/weight)+1)) #the actual formula to calculate the expected probability

    home_team_prop = np.round(we,3)
    away_team_prob = 1-home_team_prop
    away_team_prob = np.round(away_team_prob,3)
    return home_team_prop, away_team_prob

def calculate_elo(elo_home, elo_away, final_result, k_value = 40, weight = 400, home_field_advantage = 100):
    k=k_value

    erh, era = expected_result_prob(elo_home,elo_away, weight = weight, home_field_advantage = home_field_advantage)
    arh, ara = actual_result(final_result)

    # the actual formula to calculate the new elo
    updated_elo_home=elo_home + k*(arh-erh)
    updated_elo_away=elo_away + k*(ara-era)

    # round to nearest integer
    updated_elo_home=np.round(updated_elo_home, 3)
    updated_elo_away=np.round(updated_elo_away, 3)


    return updated_elo_home, updated_elo_away

def actual_result(result):
    if result == 'H':
        home_team=1
        away_team=0
    elif result == 'A':
        home_team=0
        away_team=1
    elif result == 'D':
        home_team=0.5
        away_team=0.5
    return home_team, away_team


elo_ratings = {}
df['HomeTeamELO'] = 0
df['AwayTeamELO'] = 0

# Iterate through each row (match) in the DataFrame
for index, row in df.iterrows():
    
    date = row['Date']
    home_team_id = row['HomeTeamId']
    away_team_id = row['AwayTeamId']
    home_goals = row['Home_Goals']
    away_goals = row['Away_Goals']
    margin = np.abs(home_goals - away_goals)
    if home_goals > away_goals:
        result = 'H'
    elif away_goals > home_goals:
        result = 'A'
    else:
        result = 'D'

    # Check if teams have Elo ratings, if not, initialize with the default value
    if home_team_id not in elo_ratings:
        elo_ratings[home_team_id] = initial_elo
    if away_team_id not in elo_ratings:
        elo_ratings[away_team_id] = initial_elo
        
    # Assign Elo ratings to the DataFrame
    df.at[index, 'HomeTeamELO'] = elo_ratings[home_team_id]
    df.at[index, 'AwayTeamELO'] = elo_ratings[away_team_id]
        
    k = 40
    if margin == 2:
        k = k*1.5
    elif margin == 3:
        k = k*1.75
    elif margin >= 4:
        k = k + (0.75 + ((margin-3)/8))*k
        

    

    # Update Elo ratings based on the match result
    if home_goals > away_goals:
        elo_ratings[home_team_id], elo_ratings[away_team_id] = calculate_elo(elo_ratings[home_team_id], elo_ratings[away_team_id], result, k_value=k)
    elif away_goals > home_goals:
        elo_ratings[away_team_id], elo_ratings[home_team_id] = calculate_elo(elo_ratings[away_team_id], elo_ratings[home_team_id], result, k_value=k)
    else:  # Draw
        elo_ratings[home_team_id], elo_ratings[away_team_id] = calculate_elo(elo_ratings[home_team_id], elo_ratings[away_team_id], result, k_value=k)

df['Skill_Gap'] = np.abs(df['HomeTeamELO'] - df['AwayTeamELO'] + 100)

df.dropna(inplace=True)
```

## Fitting the Models

```python
import statsmodels.api as sm

# Assuming your DataFrame is named df
# Select relevant columns as explanatory variables and the target variable
explanatory_vars = ['Mean_Home_Corners_Prev5', 'Mean_Away_Corners_Prev5', 'League_Avg_Corners']
target_var = 'Total_Corners'

# Add a constant term for the Poisson model
df['const'] = 1

# Fit a geometric Poisson model
geometric_poisson_model = sm.GLM(df[target_var], df[explanatory_vars], family=sm.families.Poisson(link=sm.families.links.log()))
geometric_poisson_results = geometric_poisson_model.fit()

# Fit a Poisson model
poisson_model = sm.GLM(df[target_var], df[['const'] + explanatory_vars], family=sm.families.Poisson(link=sm.families.links.log()))
poisson_results = poisson_model.fit()

# Fit a negative binomial model
neg_binomial_model = sm.GLM(df[target_var], df[['const'] + explanatory_vars], family=sm.families.NegativeBinomial())
neg_binomial_results = neg_binomial_model.fit()

# Display the results
print("Geometric Poisson Model:")
print(geometric_poisson_results.summary())

print("\nPoisson Model:")
print(poisson_results.summary())

print("\nNegative Binomial Model:")
print(neg_binomial_results.summary())
estimated_alpha = neg_binomial_results.scale
print(f"Estimated Dispersion Parameter (alpha): {estimated_alpha}")
```

```python
from scipy.stats import nbinom
from scipy.stats import poisson

# Add a constant term to the test data
test_df['const'] = 1

test_data = test_df[['const'] + explanatory_vars]


# Choose the count value for which you want the probability
chosen_count = 10  # Replace with your desired count value

# Use the fitted model to predict expected counts
expected_counts = poisson_results.predict(test_data)

print(expected_counts)
test_df['Expected_Corners'] = expected_counts

for i, row in test_df.iterrows():
    line = row['Line']
    at = 0
 
    if (line*2) % 2 == 0:
        at = poisson.pmf(line, expected_counts[i])
        under = poisson.cdf(line, expected_counts[i]) - at
        over = 1 - at - under

    else:
        under = poisson.cdf(line - 0.5, expected_counts[i])
        over = 1 - under
    
    test_df.at[i, 'P(Under)'] = under
    test_df.at[i, 'P(At)'] = at
    test_df.at[i, 'P(Over)'] = over
    
test_df['EV(Over)'] = (test_df['P(Over)']*test_df['Over'] + test_df['P(At)'])-1
test_df['EV(Under)'] = (test_df['P(Under)']*test_df['Under'] + test_df['P(At)'])-1
test_df['MaxEV'] = test_df[['EV(Under)', 'EV(Over)']].max(axis=1)
test_df['O/U'] = np.where(test_df['EV(Over)'] > test_df['EV(Under)'], 'Over', 'Under')

# Now, 'O/U' column contains 'Over' if EV(Over) > EV(Under) and 'Under' otherwise


print('There are {} matches where an expected value of >0 is attained'.format(len(test_df[test_df['MaxEV'] > 0])))
print('There are {} matches where an expected value of >0.1 is attained'.format(len(test_df[test_df['MaxEV'] > 0.1])))
print('So we will stake 1 unit in each of the matches where EV>0 and stake an additional 1.125 units in matches where EV>0.1')

for i, row in test_df.iterrows():
    if row['MaxEV'] > 0.1:
        
        stake = 2.125
        test_df.at[i, 'Bet (U/O)'] = row['O/U']
        test_df.at[i, 'Stake'] = stake
    elif row['MaxEV'] > 0:
        stake = 1
        test_df.at[i, 'Bet (U/O)'] = row['O/U']
        test_df.at[i, 'Stake'] = stake
    else:
        stake = 0
        test_df.at[i, 'Bet (U/O)'] = 'No Bet'
        test_df.at[i, 'Stake'] = stake

        
        
    
```
